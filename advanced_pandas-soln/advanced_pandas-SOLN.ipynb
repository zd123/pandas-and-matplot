```python
#### PART 0:
### GETTING THE DATA FILES WE NEED,
### ONLY TAKING YEARS > 1980
### AND MERGING THEM TOGETHER

######################################################
########## FORMATTING THE BATTING CSV FILE ###########
######################################################
batting = pd.read_csv('data/baseball-csvs/Batting.csv')

    ### JUST TAKE THE DATES WE ARE INTERESTED IN
batting = batting[ (batting['yearID'] >= 1980) & (batting['yearID'] <= 2010)]


    ### MAKING SURE I AM CALCULATING YEARLY TEAM STATS CORRECTLY
    ### BY SEEING IF MY RESULTS ARE THE SAME HAS HERE 
    ### http://www.baseball-reference.com/teams/NYY/batteam.shtml
testing = batting[ (batting['teamID'] == 'NYA') & (batting['yearID'] == 2010)]
testme = testing.sum()
    ### THEY ARE JUST FINE NOW MOVING FOWARD...

    
    ### NOW WE NEED TO CALCUATE THE TEAMS YEARLY STATS FOR EACH TEAM FOR EACH YEAR
    ### SINCE WE HAVE DATA FOR VERY YEAR FOR EVERY PLAYER FOR EVERY TEAM
    ### WE WANT TO SUM EACH PLAYERS STATS FOR EACH YEAR FOR GROUPED BY TEAM
    ### TO DO SO CORRECTLY, WE MUST FIRST GROUP BY YEARID THEN TEAMID... JUST THINK ABOUT IT
bat_grouped = batting.groupby( ['yearID', 'teamID'] ).sum()

    ### NOW, LETS JUST TAKE THE THINGS WE CARE ABOUT
    ### [u'AB', u'R', u'H', u'2B', u'3B', u'HR', u'RBI', u'SB', u'CS', u'BB', u'SO', u'SF', u'HBP']
bat_grouped = bat_grouped[  [u'AB', u'R', u'H', u'2B', u'3B', u'HR', u'RBI', u'SB', u'CS', u'BB', u'SO', u'SF', u'HBP' ] ]

    ### NOW, RESET THAT MULIT INDEX, SO THE YEARID AND TEAMID INDEX ARE NOW COLUMNS
bat_grouped = bat_grouped.reset_index()


######################################################
########## FORMATTING THE TEAMS CSV FILE #############
######################################################
teams = pd.read_csv('data/baseball-csvs/Teams.csv')

    ### JUST TAKE THE DATES WE ARE INTERESTED IN
teams = teams[ (teams['yearID'] >= 1980) & (teams['yearID'] <= 2010) ]

    ### ONLY GETTING THE STATS WE CARE ABOUT 
teams = teams[['yearID', 'teamID', 'G', 'W', 'L', 'Rank']]
teams = teams.reset_index()
del teams['index']

### SANITY CHECK
# print teams [teams['teamID'] == 'NYA']
### SANITY CHECK
# print bat_grouped[bat_grouped['teamID'] == 'NYA']

######################################################
########## MERGING THE TEAMS AND BATTING DATAFRAMES #############
######################################################

merged = pd.merge(left = teams, right = bat_grouped, on = ['teamID', 'yearID'] )
merged.head()
```

```python
######################################################
########## PART 1: FEATURE ENGENIEERING #############
######################################################

print merged.columns

# batting average
# BA = HITS / AT BATS

merged['BA'] =  merged['H'] / merged['AB'] 

### ON BASE PERCENTAGE
# OBP = HITS + BASE BY BALLS + HIT BY PITCH /  (AT BATS + BASE BY BALLS + HIT BY PITCH + SACRAFICED FLY )
# OBP = H + BB + HBP / AB + BB + HBP + SF
merged['OBP'] = (merged['H'] + merged['BB'] + merged['HBP']) /  (merged['AB'] + merged['BB'] + merged['HBP'] + merged['SF'])


### SLUGGING AVERAGE SLG
# before we do this wee need to calculate singles, which hopefull is jsut hits - sum(2b, 3b, hr)
merged['1B'] = merged['H'] - (merged['2B'] + merged['3B'] + merged['HR'] )


merged['SLG'] = ( (1 * merged['1B']) + (2 * merged['2B']) + (3 * merged['3B']) + (4 * merged['HR']) ) / merged['AB']
merged.head()

```

```python
### PART 3
#Now it is time to see the effect (affect?) Moneyball has had on players salaries since it became popular.  You will need to find which table the salary information is in, and how to join it with the batting data.
#1.  Find if the average salary of players with high, med, and low OBP.
#2.  Create a line graph for each of these groups where the x-axis is time, and the y-axis is salary.  Each group should be a different colored line all on the same graph.  

batting = pd.read_csv('data/baseball-csvs/Batting.csv')
batting = batting[ (batting['yearID'] >= 1985) & (batting['yearID'] <= 2013)]


# BATTING AVERAGE
batting['BA'] =  batting['H'] / batting['AB'] 
# ON BASE PERCENTAGE
batting['OBP'] = (batting['H'] + batting['BB'] + batting['HBP']) /  (batting['AB'] + batting['BB'] + batting['HBP'] + batting['SF'])
# SLUGGING 
batting['1B'] = batting['H'] - (batting['2B'] + batting['3B'] + batting['HR'] )
batting['SLG'] = ( (1 * batting['1B']) + (2 * batting['2B']) + (3 * batting['3B']) + (4 * batting['HR']) ) / batting['AB']
```

```python
### PART THREE QUESTION 1
#1.  Find if the average salary of players with high, med, and low OBP.


### NOW WE ARE BINNING EACH OBP BY HIGH MED AND LOW PER YEAR
### FIRST LETS CROP OOUT THE OUTLIERS, IE OBP IS 0 OR 1
### FIRST REMOVE ALL THE POEPLE THAT HAVE HAD LESS THAN 30 AT BATS

batting = batting[ batting['AB'] > 30 ]
### THEN REMOVE ALL THE PPEOPLE THAT OBP THAT OBVI DONT REFLECT A NORMAL OBP AND ARE SPECIAL CASES (PITCHERS HITTING)
batting = batting[ (batting['OBP'] > 0.01) & (batting['OBP'] < 0.75) ]


### NOW WE WANT TO MAKE THREE GROUPS OF OUR OBP FEATURE
# use pd.cut([0,5,10]) http://pandas.pydata.org/pandas-docs/stable/generated/pandas.cut.html
### PLAYERS WITH LOW, MED, AND HIGH OBP

### I DEFINE LOW BY BEING BETWEEN THE MIN AND MEAN - 1STD
### I DEFINE MID BY BEING BETWEEN MEAN-1STD AND MEAN+1STD
### I DEFINE HIGH BY BEING BEING ABOVE THE MEAN+1STD AND THE MAX
info = batting['OBP'].describe()
mean = info['mean']
std = info['std']

lowest = info['min']
low = mean - std
high = mean + std
highest = info['max']

bins = [lowest, low, high, highest]
batting['bins'] = pd.cut(batting['OBP'], bins)

#### NOW THAT I HAVE MY BINS, LETS INSERT THE SALARY DATA FOR ALL THE PLAYERS
sals = pd.read_csv('data/baseball-csvs/Salaries.csv')
sals = sals[ sals['yearID'] >= 1980 ]   #### sals = sals[ (sals['yearID'] >= 1985) & (sals['yearID' <= 2013]) ]
sals = sals[['yearID', 'playerID', 'salary']]
mdf = pd.merge(batting, sals, on=['playerID', 'yearID'] )


### NOW THAT OUR SALARY DATA IS IN THERE, LETS GROUP BY BIN AND FIND THE AVERAGE SALARY FOR EACH BIN

#### ANSWER FOR PART 3 Q1 
mdf.groupby('bins')['salary'].mean().plot(kind='bar')
mdf.groupby('bins')['salary'].mean()
```

```python
#### PART 3 QUESTION 2:
#2.  Create a line graph for each of these groups where the x-axis is time, and the y-axis is salary.  Each group should be a different colored line all on the same graph.  

### NOW LESTS SEE HOW THIS HAS CHANGED OVER THE YEARS
peryear = mdf.groupby( ['yearID', 'bins'])['salary'].mean()
peryear = peryear.reset_index()

### JUST REPLACING THE STUPD NAMES PD.CUT GIVES WITH LOW MED HIGH
peryear.bins.replace(to_replace = '(0.027, 0.237]', value='low', inplace=True )
peryear.bins.replace(to_replace = '(0.237, 0.373]', value='mid', inplace=True )
peryear.bins.replace(to_replace = '(0.373, 0.609]', value='high', inplace=True )
peryear.plot(kind='bar')

# lowz = peryear[peryear.bins == 'low']
# midz = peryear[peryear.bins == 'mid']
# highz = peryear[peryear.bins == 'high']

# # peryear = peryear.set_index('yearID')
# peryear.head()
# # peryear.plot(x='yearID', y='salary', )
*** STOPPED HERE BECAUSE CANNOT FIGURE OUT HOW TO GROUP AND PLOT EACH BIN AS ITS OWN COLOR LINE
```
