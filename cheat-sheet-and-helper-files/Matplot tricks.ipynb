````python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
````

````python
x1 = np.linspace(0.0, 5.0)
x2 = np.linspace(0.0, 2.0)

y1 = np.cos(2 * np.pi * x1) * np.exp(-x1)
y2 = np.cos(2 * np.pi * x2)

plt.subplot(3, 1, 1)
plt.plot(x1, y1, 'yo-')
plt.title('A tale of 2 subplots')
plt.ylabel('Damped oscillation')

plt.subplot(3, 1, 2)
plt.plot(x2, y2, 'r.-')
plt.xlabel('time (s)')
plt.ylabel('Undamped')

plt.show()
plt.subplot(3, 1, 3)
plt.plot(x1, y1, color='green', linestyle='dashed', marker='o',
     markerfacecolor='blue', markersize=4)
````

````python
### ONCE YOU AHVE ESTABLISHED YOUR FIGURE OBJECT, 
### NOW YOU CAN ADD CHARTS TO TI, YOU DO 
### fig.add_subplot()

fig = plt.figure()
chart = fig.add_subplot(1, 1, 1)
chart.plot()
````

````python
#### NOW LETS ADD DATA TO THE PLOT 
data = np.arange(0,100,1)

fig = plt.figure()
plt.xlabel('x')
plt.ylabel('y')
plt.title('title')

chart = fig.add_subplot(1, 1, 1)
chart.plot( data ) 
````

````python
#### LETS SPECIFIY OUR X AND Y DATA 
data_x = np.arange(0,100,1)
data_y = np.random.normal(size=100)

fig = plt.figure()
chart = fig.add_subplot(1, 1, 1)
chart.plot( data_x,  data_y )
````

````python
#### IF WE WANT TO PLOT MORE THAN ONE LINE, JUST CALL THE CHART OBJECT AGAIN, AND ADD MORE DATA

data_x = np.arange(0,100,1)
data_y1 = np.random.normal(size=100)
data_y2 = np.random.poisson(size=100)

fig = plt.figure()

chart = fig.add_subplot(1, 1, 1)
chart.plot( data_x,  data_y1 )
chart.plot( data_x,  data_y2 )
````

* The third one chanages
